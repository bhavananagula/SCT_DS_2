```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from google.colab import files

#  Upload dataset
uploaded = files.upload()
file = list(uploaded.keys())[0]

# Load dataset
df = pd.read_csv(file)
print("Dataset loaded successfully!\n")
print("Columns available:\n", df.columns, "\n")
print(df.head())


#  Standard Titanic dataset columns: Survived, Pclass, Sex, Age, SibSp, Parch, Fare, Embarked

# If "Age" exists → fill missing with median
if "Age" in df.columns:
    df["Age"].fillna(df["Age"].median(), inplace=True)

    # If "Embarked" exists → fill missing with mode
    if "Embarked" in df.columns:
        df["Embarked"].fillna(df["Embarked"].mode()[0], inplace=True)

        # If "Cabin" exists → drop (too many missing values)
        if "Cabin" in df.columns:
            df.drop(columns=["Cabin"], inplace=True)

            # Drop PassengerId, Name, Ticket (not useful for EDA)
            for col in ["PassengerId", "Name", "Ticket"]:
                if col in df.columns:
                        df.drop(columns=[col], inplace=True)

                        print("\n✅ Data cleaned successfully!")
                        print(df.isnull().sum())


#  Survival distribution
if "Survived" in df.columns:
    sns.countplot(x="Survived", data=df, palette="Set2")
    plt.title("Survival Count")
    plt.show()

    #  Gender vs Survival
    if "Sex" in df.columns and "Survived" in df.columns:
        sns.countplot(x="Sex", hue="Survived", data=df, palette="Set1")
        plt.title("Survival by Gender")
        plt.show()

        #  Age distribution
        if "Age" in df.columns:
            plt.figure(figsize=(10,6))
            sns.histplot(df["Age"], bins=30, kde=True, color="skyblue")
            plt.title("Age Distribution")
            plt.show()

            # Passenger Class vs Survival
            if "Pclass" in df.columns and "Survived" in df.columns:
                sns.countplot(x="Pclass", hue="Survived", data=df, palette="Set3")
                plt.title("Survival by Passenger Class")
                plt.show()

                # Correlation Heatmap
                plt.figure(figsize=(8,6))
                sns.heatmap(df.corr(), annot=True, cmap="coolwarm")
                plt.title("Correlation Heatmap")
                plt.show()
```
